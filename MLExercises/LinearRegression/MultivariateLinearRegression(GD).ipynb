```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("ex1data2.txt",header=None)
```

```python
def computeCost(X,y,theta):
    
    m = len(y)
    h = X.dot(theta)
    square_error = (h - y) **2
    return 1/(2*m) * np.sum(square_error)
```

```python
#Normalize features using mean normalization
def featureNormalize(X):
    
    mu = np.mean(X)
    print(mu)
    sigma = np.std(X)
    
    a = X-mu
    X_norm = a/sigma
    return X_norm
    
```

```python
data_n=data.values                   #convert to array
m=len(data_n[:,0])
 
X=np.array(data_n[:,0])
X=np.append(X.reshape(m,1),data_n[:,1].reshape(m,1),axis=1)

print(X.shape)
#X=np.append(np.ones((m,1)),data_n[:,0].reshape(m,1),axis=1)
X =featureNormalize(X)
print(X.shape)
X=np.append(np.ones((m,1)),X.reshape(m,2),axis=1)
print(X.shape)

Y=data_n[:,2].reshape(m,1)

theta=np.zeros((3,1))

computeCost(X,Y,theta)
print(Y)
```

```python
#gradient descent
def gradientDescent(X,y,theta,alpha,num_iters):
    
    m = len(y)
    J_history = []
    
    for i in range(num_iters):
        h = X.dot(theta)
        error = np.dot(X.transpose(),(h-y))   #summation(h-y)*X
        print(error)
        descent = alpha * 1/m *error
        theta -= descent
        J_history.append(computeCost(X,y,theta))
        
    return theta,J_history
```

```python
theta,J_history = gradientDescent(X,Y,theta,0.01,400)
print("h(x) = "+ str(round(theta[0,0],2))+"+"+str(round(theta[1,0],2))+"x1")
```

```python
#Predict values for population sizes of 35,000 and 70,000
array = np.array([1, 1650,3])
prediction = theta.transpose().dot(array)
print("Predicted price for population size of 35,000 = ",prediction * 10000)

```
